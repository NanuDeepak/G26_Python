K MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

```python
# printing all data types (for converting object types)
for column in data.columns:
  if data[column].dtype == object:
    print(str(column) + ' : '+str(data[column].unique()))
    print(data[column].value_counts())
    print('_______')
```

```python
#Transform object into numerical col
from sklearn.preprocessing import LabelEncoder

for column in data.columns:
  if data[column].dtype == np. number:
    continue
  data[column] = LabelEncoder().fit_transform(data[column])
```

```python
data
```

```python
from pandas.core.internals.blocks import NumericBlock
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(6,2.5))
plt.plot(range(1,10),inertia,"*r-")
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmean = KMeans(n_clusters=2)
predict = kmean.fit_predict(data)
```

```python
x = data
```

```python
plt.figure(figsize=(6,2.5))
plt.scatter(data['Age'], data['Attrition'], c='b', marker='o', label='Data Points')

data['kmeans_cluster'] = kmeans.fit_predict(x)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black')

plt.xlabel('Age')
plt.ylabel('Attrition')
plt.title('Data Distribution')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,2.5))
plt.scatter(x.iloc[predict == 0,0],x.iloc[predict == 0,1],c="r")
plt.scatter(x.iloc[predict == 1,0],x.iloc[predict == 1,1],c="b")
```

```python
plt.scatter(data['Age'], data['HourlyRate'], c='b', marker='o', label='Data Points')

data['kmeans_cluster'] = kmeans.fit_predict(x)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black')

plt.xlabel('Age')
plt.ylabel('HourlyRate')
plt.title('Data Distribution')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python

features = data[['Age', 'DistanceFromHome', 'Education', 'HourlyRate', 'JobLevel']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)


plt.scatter(data['Age'], data['HourlyRate'], c=data['Cluster'], cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('HourlyRate')
plt.title('Employee Performance Clustering')
plt.legend()
plt.show()
```

```python

features = data[['JobLevel', 'HourlyRate']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)


plt.scatter(data['JobLevel'], data['HourlyRate'], c=data['Cluster'], cmap='rainbow')
plt.xlabel('JobLevel')
plt.ylabel('HourlyRate')
plt.title('Employee Performance Clustering')
plt.legend()
plt.show()
```

**As the Level 1 are senior level employees they will get paid more .Above plot shoes the same the level 1 (senior employees ) get paid a lot**
