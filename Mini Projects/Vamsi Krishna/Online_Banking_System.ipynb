# Online Banking System

```python
# userdata base of UserDataBase works like databasing unit
UserDataBase = {
    "user1": {"password": "password1", "balance": 10000},
    "user2": {"password": "password2", "balance": 2500},
    "user3": {"password": "password3", "balance": 12500},
}


#login unit
def login():
    username = input("Username: ") #input from the user for username
    password = input("Password: ") #input from the user for password

    if username in UserDataBase and password in UserDataBase == username or password:   # UserDataBase[username]["password"]   #it will checks the entered username and password is exists in dictionary UserDataBase
        print("<......................Login successful!...............................>")
        return username
    else:
        print("<................Invalid credentials. Please try again..................>")
        return None


#check_balance unit
def check_balance(username): #view balnace changed to check balnce
    balance = UserDataBase[username]["balance"]
    print(f"Your account balance is: {balance}") #here f is used here for string holder


#transaction unit
def do_transaction(username): #make_transaction changed to do transaction
    print("....>>>>  Account Transaction <<<<....")
    amount = float(input("Enter the transaction amount: "))
    transaction_type = input("Enter '1' for deposit or '2' for withdrawal: ")

    if transaction_type == "1":
        print(" ...............>>>>  You choosed Deposit Option   <<<<...............")
        UserDataBase[username]["balance"] += amount
        print(f"Deposit successful. Your new balance is: {UserDataBase[username]['balance']}")
    elif transaction_type == "2":
        print(" ...............>>>>  You choosed Withdrawl Option   <<<<...............")
        if amount > UserDataBase[username]["balance"]:
            print("Insufficient funds. Transaction canceled.")
        else:
            UserDataBase[username]["balance"] -= amount
            print(f"Withdrawal successful. Your new balance is: {UserDataBase[username]['balance']}")
    else:
        print("Invalid transaction type. Please enter '1' or '2'.")


def main():
    print(" .............................>>>>>  Welcome  <<<<<........................... ")
    print("              ............>>>>>  Online Banking System  <<<<<...........       ")
    print("............>>>>>  Enter your username and password to log in  <<<<<...........")
    username = None

    while not username:
        username = login()

    while True:
        print("\n \n")
        print("Options:")
        print("1. Check Balance")
        print("2. Make a Transaction")
        print("3. Log Out")
        choice = int(input("Enter your choice from the above (1/2/3): "))

        if choice == 1:
            check_balance(username)
        elif choice == 2:
            do_transaction(username)
        elif choice == 3:
            print("...>>>> Thank you for using our Online Banking System  <<<<....")
            print("...>>>>   .Logging out.   <<<<...")
            break
        else:
            print("Invalid choice. Please try again.")

main()

```
